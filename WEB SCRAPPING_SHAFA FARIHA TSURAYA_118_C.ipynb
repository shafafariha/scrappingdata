```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://top-1000-sekolah.ltmpt.ac.id/'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table', {'class': 'table table-striped table-bordered'})

peringkat_sekarang = []
peringkat_sebelumnya = []
npsn = []
sekolah = []
nilai_total = []
provinsi = []
kabupaten_kota = []
jenis_sekolah = []

rows = table.find_all('tr')[1:]

for row in rows:
    columns = row.find_all('td')
    peringkat_sekarang.append(columns[0].text)
    peringkat_sebelumnya.append(columns[1].text)
    npsn.append(columns[2].text)
    sekolah.append(columns[3].text)
    nilai_total.append(columns[4].text)
    provinsi.append(columns[5].text)
    kabupaten_kota.append(columns[6].text)
    jenis_sekolah.append(columns[7].text)

data = {
    "Peringkat Sekarang": peringkat_sekarang,
    "Peringkat Sebelumnya": peringkat_sebelumnya,
    "NPSN": npsn,
    "Nama Sekolah": sekolah,
    "Total Nilai": nilai_total,
    "Provinsi": provinsi,
    "Kabupaten/Kota": kabupaten_kota,
    "Jenis Sekolah": jenis_sekolah
}
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv('output.csv', index=False)
```
